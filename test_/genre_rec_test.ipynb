```python
#final function

def recommend_by_genre(liked_movies, top_n=5):

    # Load movie dataset
    df_imbd = pd.read_csv(IMDB_PATH)

    # safety check — no input movies
    if not liked_movies:
        return []

    #  Task 1:
    #  we get the user liked movies
    #  we visit each movie using 'Series_Title'
    #  and then based on that movie we visit the 'Genre' columns and extract all the Genre

    liked_genre_temp = []

    for title in liked_movies:
        # grab the current movie row
        row = df_imbd[df_imbd["Series_Title"] == title]

        # safety check — movie not found
        if row.empty:
            continue

        # now get the genre index content
        genre = row.iloc[0]["Genre"]

        # now split the each genre
        genre = genre.split(",")

        # remove the leading spaces of each genre and append to liked genre list
        for g in genre:
            liked_genre_temp.append(g.strip())

    # safety check — no genres extracted
    if not liked_genre_temp:
        return []

    # task 2 based on the genre frequency give the score to each genre
    genre_score = {}

    for g in liked_genre_temp:
        genre_score[g] = genre_score.get(g, 0) + 1  

    recommendations = []

    # then based upon the score we will visit new movies which is not watched by the user
    for index, row in df_imbd.iterrows():
        title = row["Series_Title"]
        imbd_rating = row["IMDB_Rating"]
        

        # don't recommend already liked movies
        if title in liked_movies:
            continue

        genres = row["Genre"].split(",")
        movie_genres = []

        for g in genres:
            movie_genres.append(g.strip())

        # calculating score for this current movie
        score = 0
        for g in movie_genres:
            score += genre_score.get(g, 0)

        if score > 0:
            recommendations.append((title, score,imbd_rating))

    # now sort the movies based on the score and imbd rating
    recommendations.sort(key=lambda i: (i[1],i[2]), reverse=True)

    # return only top N movie names
    recommendations_movies_name = []

    for title, score ,rating in recommendations[:top_n]:
        recommendations_movies_name.append(title)

    return recommendations_movies_name
```
